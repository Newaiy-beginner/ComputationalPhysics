```python
import numpy as np
import numpy.linalg as LA
import matplotlib.pyplot as plt
import math

def V(x, l, v):
    v = math.sqrt(76.2*((2*l)**2)/v)
    if x < -l:
        return v
    elif x > l:
        return v
    else:
        return 0.0

l = 3.0
hbar = 1
mass= 1
v = 900
print(V(2*l, l, v))

L = 3*l
# Given N, L
N = 200
N = 2*N+1   # odd num of grids
x = np.linspace(-L, L, N)

H = np.zeros((N, N), float)
h = 2*L/(N-1)
print(h)
# Build H:
H += np.eye(N, k = -1)
H += np.eye(N, k = 1)
# Finite difference
H *= 1/(h**2)
H += -(2/h**2)*np.eye(N, k = 0)
H *= -0.5   # Schrodinger
for i in range(N):
    H[i][i] += V(x[i], l, v)
val, vec = LA.eig(H)
idx = np.argsort(val)
val = val[idx]
vec = vec[:, idx]

for i in range(0, 5):
    if i%2 == 0:
        plt.plot(x, val[i] + 20*vec[:, i]**2, 'r-')
    else:
        plt.plot(x, val[i] + 20*vec[:, i]**2, 'b-.')
plt.ylim(0, val[10]+0.1)
#plt.ylim(7.75, 8.0)
plt.axvline(-l, 0.0, 1.0)
plt.axvline(l, 0.0, 1.0)
plt.axhline(math.sqrt(76.2*((2*l)**2)/v))
plt.title("Intensity(x20); width = "+str(2*l)+"nm")
plt.show()
#plt.savefig("1D_SHO.png")

# Analytic solution of energy eigenvalue
def E_inf(n):
    return (hbar**2)/(2.0*mass)*((n+1)*math.pi/(2.0*l))**2.0

xEnergy = np.arange(0, 6, 1)
plt.plot(xEnergy, 76.2*val[0:len(xEnergy)], 'b.', zorder = 3)
n = np.linspace(0, 5, 50)
plt.plot(n, 76.2*E_inf(n), 'r-')
plt.xlabel("n")
plt.ylabel("Energy(meV)")
plt.grid(True)
plt.show()




# HW: take square root to check the answer
# L enough large
```
