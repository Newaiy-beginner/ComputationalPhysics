```python
import numpy as np
import matplotlib.pyplot as plt
import math

N = 10
xi = []
fi = []

for i in range(N):
    xi.append(2.0*i*math.pi/10.0)
    fi.append(math.cos(xi[i]))
'''
for i in range(N):
    xi.append(i*25)
fi = [10.6, 16.0, 45.0, 83.5, 52.8, 19.9, 10.8, 8.25, 4.7]
'''
def P(x, xk, j, n):
    p = 1.0
    for k in range(n):
        if k != j:
            p *= (x-xk[k])/(xk[j]-xk[k])
    return p

def f_lagrange(x, xk, fk, n):
    s = 0.0
    for j in range(n):
        s += P(x, xk, j, N)*fk[j]
    return s

x_real = np.linspace(0, 2*math.pi, 100)
f_real = np.zeros(100)
f_anal = np.zeros(100)

for i in range(100):
    f_real[i]=f_lagrange(x_real[i], xi, fi, N)
    f_anal[i]=math.cos(x_real[i])
plt.grid(True)
plt.plot(x_real, f_real)
plt.plot(xi, fi, "r.")
plt.plot(x_real, f_anal, "g-.", zorder = 5)
plt.show()
```
