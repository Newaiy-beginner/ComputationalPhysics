```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA
import math

def FDM_UniformChargedLine(N, case):
    Size = N**2
    A = np.eye(Size, dtype = float)
    rho = np.zeros(Size, float)
    posiCenter = Size//2 - N*5
    negaCenter = Size//2 + N*5
    if case == 1:
        # Square distribution of electrons
        m = 5
        for i in range(m):
            rho[(Size//2+(i-m//2)*N-m//2):(Size//2+(i-m//2)*N+m//2+1)] = 1.0
    else:
        # 2D uniform distribution of electrons
        rho[posiCenter-3: posiCenter+4] = 1.0
        rho[negaCenter-3: negaCenter+4] = -1.0
    # Build coefficient matrix of potential
    i = N
    while i < Size-N:
        if (i-0)%N == 0 or (i-(N-1))%N == 0:
            i += 1
        else:
            A[i][i] = -4.0
            A[i][i-1] = 1.0
            A[i][i+1] = 1.0
            A[i][i-N] = 1.0
            A[i][i+N] = 1.0
            i += 1
    #print(A)
    V = LA.inv(A).dot(rho)
    #print(V)
    Vmat = np.zeros((N, N), float)
    for i in range(N):
        Vmat[i] = V[(i*N):(i*N + N)]  
    return -1*Vmat

def Efield(N, case):
    # Electric field
    V = FDM_UniformChargedLine(N, case)
    # N-1 by N-1 matrix
    Esize = N - 1
    Ex = np.zeros((Esize, Esize), float)
    Ey = np.zeros((Esize, Esize), float)
    for i in range(Esize):
        for j in range(Esize):
            # Differential in two directions
            Ex[i][j] = (-0.5)*(V[i+1][j+1] - V[i+1][j] + V[i][j+1] - V[i][j])
            Ey[i][j] = (-0.5)*(V[i+1][j+1] - V[i][j+1] + V[i+1][j] - V[i][j])
    return [Ex, Ey]

N = 21
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

X = np.linspace(1, N, N)
Y = np.linspace(1, N, N)
X, Y = np.meshgrid(X, Y)
plt.title(N)
plt.contourf(X, Y, FDM_UniformChargedLine(N, 1), 7)

E = Efield(N, 1)
X = np.zeros(N-1, float)
Y = np.zeros(N-1, float)
for i in range(len(X)):
    X[i] = i+1.5
    Y[i] = i+1.5
X, Y = np.meshgrid(X, Y)
U = E[0]
V = E[1]
q = ax.quiver(X, Y, U, V, scale = 12.4)
plt.show()
'''
# plot
plt.figure()
N = [3]
for i in range(len(N)):
    X = np.linspace(0, N[i], N[i])
    Y = np.linspace(0, N[i], N[i])
    X, Y = np.meshgrid(X, Y)
    #plt.subplot(2, 2, i+1)
    plt.title(N[i])
    plt.contourf(X, Y, FDM_UniformChargedLine(N[i]), 4)
plt.show()
'''
```
