```python
import numpy as np
import matplotlib.pyplot as plt
import math
import cmath

delta = [0.0, 0.5, 1.0, 3.0, 3.5, 5.0]; delta = [0.01*delta[i] for i in range(len(delta))]
omega0 = 100.0
omega = [omega0*(1+delta[i]) for i in range(len(delta))]
rabiFreq = 0.01*omega0

def g1(t, y1, y2, idx):
    return 0.5*1j*rabiFreq*(cmath.exp(1j*(omega[idx]-omega0)*t) + cmath.exp(-1j*(omega[idx]+omega0)*t))*y2
def g2(t, y1, y2, idx):
    return 0.5*1j*rabiFreq*(cmath.exp(-1j*(omega[idx]-omega0)*t) + cmath.exp(1j*(omega[idx]+omega0)*t))*y1

def rk2(ti, h, y10, y20, w1, w2, alpha, beta, idx):
    k = g1(ti, y10, y20, idx)
    q = g2(ti, y10, y20, idx)
    k1 = g1(ti+alpha*h, y10+beta*h*k, y20+beta*h*q, idx)
    q1 = g2(ti+alpha*h, y10+beta*h*k, y20+beta*h*q, idx)
    yp1 = y10 + h*(w1*k + w2*k1)
    yp2 = y20 + h*(w1*q + w2*q1)
    return yp1, yp2

N = 7000
h = 0.001
c1 = np.zeros(N, complex)
c2 = np.zeros(N, complex)

t = 0.0 + h*np.arange(N)
c1[0] = complex(1.0, 0.0); c2[0] = math.sqrt(1-abs(c1[0])**2)

plt.figure(figsize = (14,8))
plt.grid(True)
for n in range(len(delta)):
    print("Processing "+str(n+1)+"...")
    for i in range(N-1):
        c1[i+1], c2[i+1] = rk2(t[i], h, c1[i], c2[i], 0.5, 0.5, 1.0, 1.0, n)    
    plt.plot(t, [abs(c1[i])**2 for i in range(len(c1))], label="$|C_1|^2$, $\delta=$"+str(delta[n]))
    c1[0] = complex(1.0, 0.0); c2[0] = math.sqrt(1-abs(c1[0])**2)
    # plt.plot(t, [abs(c2[i])**2 for i in range(len(c2))])
plt.legend()
```
