```python
import numpy as np
import numpy.linalg as LA
import matplotlib.pyplot as plt
import math

e = 1.0
E = 1.0  # 
# ('Eunit') meV
hbar = 1.0; mass = 1.0  #
hbarConst = 6.62607015*(10**(-34))/(2.0*math.pi); massConst = 9.10938356*(10**(-31))  # J*sec; kg
eVconst = 1.602176634*(10**(-19))  # Joule
Eunit = (hbarConst**2)/(massConst*(10**(-18)))*1000
Eunit /= eVconst

# infinity
Bound = 10**8
L = 15.0; l = 10.0    # Unit: nm
# odd number
N = 400; N = 2*N+1
# Spacing
h = 2*L/(N-1)
print("Spacing =", h)

# field
def V(x, field):
    V0 = 0.0
    ans = e*field*x + V0
    if x <= -l/2:
        return Bound
    elif x >= l/2:
        return Bound
    else:
        return ans

#  numS  Simpson method  number of slides
def Intgral(arr, numS):
    # Simpson's method
    # Given function array, number of Slides
    s = 0.0
    for n in range(numS):
        s += (1.0/3.0)*h*(arr[2*n] + 4.0*arr[2*n] + arr[2*n])
    return s

# Finite difference  input  field ()
def Hamiltonian(field):
    xi = np.linspace(-L, L, N)
    # Build Hamiltonian
    H = np.zeros((N, N), float); H += np.eye(N, k = -1); H += np.eye(N, k = 1)
    # Finite difference
    H *= 1/(2.0*(h**2))
    H += -(1/h**2)*np.eye(N, k = 0)
    for i in range(N):
        H[i][i] -= (mass/(hbar**2.0))*V(xi[i], field)
    H *= -(hbar**2.0)/mass
    #Compute eigval and eigfunc
    val, vec = LA.eig(H)
    idx = np.argsort(val)
    val = val[idx]
    vec = vec[:, idx]

    # eigenfunctions Normalization
    intense = np.zeros(len(vec[:,3]))
    for i in range(len(intense)):
        intense[i] = abs(vec[i,3])**2.0
    C = Intgral(intense, N//2); C = 1.0/math.sqrt(C)
    vec *= C
    return xi, vec[:, 0:20], val[:20]

#  eigenfunctions 
def wavePlot(field):
    plt.figure(figsize = (11, 7))
    plt.grid(True)
    x, eigvec, eigval = Hamiltonian(field)
    eigval *= Eunit

    # Plot Potential    
    Poten = np.zeros(len(x))
    for i in range(len(Poten)):
        if x[i] >= -l/2 and x[i] <= l/2:
            Poten[i] = V(x[i],field)
        else:
            Poten[i] = 1000.0
    Poten *= Eunit
            
    # Plot wave function(probability density)
    plt.plot(x, eigval[0] + abs(10*eigvec[:, 0])**2.0, 'r-', label="$10|\psi_n|^2$")
    for i in range(1, 10):
        plt.plot(x, eigval[i] + abs(10*eigvec[:, i])**2.0, 'r-')
    plt.ylabel("$Energy(meV)$")
    plt.xlabel("$x(nm)$")

    plt.ylim(min(Poten), eigval[9]+50.0)
    plt.plot(x, Poten[:], 'b-')
    plt.xlabel("$x(nm)$")
    plt.legend()
    plt.show()
wavePlot(0.05)
```

```python
#  eigenvalue 
def eigenE_plot(field):
    plt.figure(figsize = (11, 7))
    plt.grid(True)
    x, eigvec, eigval = Hamiltonian(field)
    x, eigvec, eigval2 = Hamiltonian(0.0)
    xEnergy = np.arange(0, 19, 1)

    plt.plot(xEnergy, eigval[0:len(xEnergy)], 'bo', label="Field="+str(field))
    k = [n*math.pi/l for n in range(1, len(xEnergy)+1)]
    Eanal = [((hbar*k[n])**2.0)/(2.0*mass) for n in range(len(k))]
    plt.plot(xEnergy, eigval2[0:len(xEnergy)], 'c^', label="Field off")
    plt.plot(xEnergy, Eanal, 'r-.', label="Field off(Analytic)")
    plt.xlabel("n")
    plt.ylabel("Energy")
    plt.legend()
    plt.show()
eigenE_plot(0.1)
```

```python
#  Transition rate  F_arr  array.
def TransiPlot(F_arr):
# transition matrix element of n=1 to n = 2~9
    color = ['b', 'r', 'g', 'y']
    plt.figure(figsize = (11, 7))
    plt.grid(True)
    for w in range(len(F_arr)):
        print("Processing "+str(w+1)+"...")
        x, eigvec, eigval = Hamiltonian(F_arr[w])
        # Optical transition element
        psi1 = eigvec[:,0]    # Ground state

        prob = []
        for j in range(1,10):
            psi2 = eigvec[:, 1+j]    # (1+j) excited state
            f = np.zeros(len(psi1))
            for i in range(len(psi1)):
                f[i] = psi2[i]*(e*E*x[i])*psi1[i]
            prob.append(abs(Intgral(f, N//2))**2.0)
        plt.plot([(i) for i in range(9)], prob, color[w]+'^-', 
                 label="$|V_{n1}|^2$, F="+str(F_arr[w])+"V/nm")
    plt.xlabel("Transition States")
    plt.title("Perturb light field ="+str(E)+" Volt/nm")
    plt.legend()
    plt.show()
TransiPlot([0.00])
TransiPlot([0.00, 0.01, 0.02, 0.1])
```
