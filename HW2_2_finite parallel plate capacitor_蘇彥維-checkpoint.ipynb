```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#N = 11
N = 20
V = np.ones((N, N), float)

# Given initial value(guess)
V*=0.0

# Set position and shape of parallel capacitors
left = 5
right = N-left
V[left, left:right] = -1.0
V[(right-1), left:right] = 1.0

Vtop = 0.0
Vbottom = 0.0
Vright = 0.0
Vleft = 0.0
V[:1, :] = Vtop
V[(N-1):, :] = Vbottom
V[:, (N-1):] = Vright
V[:, :1] = Vleft
```

```python
# Charge
def rho(i, j):
    if i == left:
        if j < right and j >= left:
            return -1.0
        return 0.0
    elif i == right-1:
        if j < right and j >= left:
            return 1.0
        return 0.0
    return 0.0
# B.C. by parallel plate capacitors
def BC(i, j):
    if rho(i, j) != 0.0:
        return True
    return False

def Efield(N, v):
    # Electric field
    # N-1 by N-1 matrix
    Esize = N - 1
    Ex = np.zeros((Esize, Esize), float)
    Ey = np.zeros((Esize, Esize), float)
    for i in range(Esize):
        for j in range(Esize):
            # Differential in two directions
            Ex[i][j] = (-0.5)*(v[i+1][j+1] - v[i+1][j] + v[i][j+1] - v[i][j])
            Ey[i][j] = (-0.5)*(v[i+1][j+1] - v[i][j+1] + v[i+1][j] - v[i][j])
    return [Ex, Ey]
```

```python
times = 0
err = 1.0
n = 0
# relaxation
while err > 0.000001:
    err = 0.0
    ierr = 0
    # don't calculate the boundary(by B.C.)
    for i in range(1, N-1):
        for j in range(1, N-1):
            v0 = V[i, j]
            # B.C.
            if not BC(i, j):
                V[i, j] = 0.25*(V[i+1][j] + V[i-1][j] + V[i][j+1] + V[i][j-1] + rho(i, j))
            err += np.abs(V[i, j] - v0)
            ierr += 1
    err /= ierr
    times += 1
    #print("times", times, ":", err)
    #er = np.append(er, verr)
    #print("it=", it[n], ":verr=", n)
print("Iteration times =", times)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
X = np.linspace(1, N, N)
Y = np.linspace(1, N, N)
X, Y = np.meshgrid(X, Y)

plt.contourf(X, Y, V, 50, cmap=plt.cm.jet)
plt.colorbar()

E = Efield(N, V)
X = np.zeros(N-1, float)
Y = np.zeros(N-1, float)
for i in range(len(X)):
    X[i] = i+1.5
    Y[i] = i+1.5
X, Y = np.meshgrid(X, Y)
U = E[0]
V = E[1]
q = ax.quiver(X, Y, U, V, scale = 10.4)
#plt.savefig("hw2_2.png")
plt.show()
```
