```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Constant
ga = 9.8
k = 0.1
mass = 1.0
# ODE
def g(t, y):
    return ga - (k/mass)*y

N = 201
ti = 0.0
tf = 50.0
t = np.linspace(ti, tf, N)
h = (tf - ti)/(N-1)
print("h =", h)

yRk = np.zeros(N)
yEuler = np.zeros(N)
# Initial condition
yRk[0] = 0.0
yEuler[0] = 0.0
# Analytic solution
anal = np.zeros(N)
for i in range(N):
    anal[i] = ((mass*ga)/k)*(1.0 - math.exp(-(k/mass)*t[i]))

for i in range(0, N-1):
    # Euler method
    yEuler[i+1] = yEuler[i] + h*g(t[i], yEuler[i])
    # Rk2
    k1 = g(t[i], yRk[i])
    k2 = g(t[i]+h, yRk[i]+h*k1)
    yRk[i+1] = yRk[i] + h*(0.5*k1 + 0.5*k2)

# Error
errEuler = [abs(abs(yEuler[i]-anal[i])/yEuler[i]) for i in range(len(anal))]
errRk = [abs(abs(yRk[i]-anal[i])/yRk[i]) for i in range(len(anal))]

errEulerAbs = [abs(yEuler[i]-anal[i]) for i in range(len(anal))]
errRkAbs = [abs(yRk[i]-anal[i]) for i in range(len(anal))]

#Plot
plt.grid(True)
plt.plot(t, yRk, 'b:', zorder = 3)
plt.plot(t, anal, 'r-')
plt.show()

plt.grid(True)
plt.plot(t, errRk, 'b:')
plt.plot(t, errRkAbs, 'r:')
plt.show()
print("Absolute error converge:", errRkAbs[len(errRkAbs) - 1])
print("Relative error converge:", errRk[len(errRk) - 1])

plt.grid(True)
plt.plot(t, yEuler, 'b:', zorder = 3)
plt.plot(t, anal, 'r-')
plt.show()

plt.grid(True)
plt.plot(t, errEuler, 'b:')
plt.plot(t, errEulerAbs, 'r:')
plt.show()
print("Absolute error converge:", errEulerAbs[len(errEulerAbs) - 1])
print("Relative error converge:", errEuler[len(errEuler) - 1])
```
