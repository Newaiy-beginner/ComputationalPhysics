```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 40
V = np.zeros((N, N), float)

# Given initial value(guess)
V*=0.15

# Set position and shape of parallel capacitors
left = 4
right = left + 10
V[N//2, left:right] = 1.0
V[N//2, (N-right):(N-left)] = 1.0

Vtop = 0.0
Vbottom = 0.0
Vright = 0.0
Vleft = 0.0
V[:1, :] = Vtop
V[(N-1):, :] = Vbottom
V[:, (N-1):] = Vright
V[:, :1] = Vleft
```

```python
# B.C. by parallel plate capacitors
def BC(i, j):
    if i == N//2:
        if j >= left and j < right:
            return True
        elif j >= (N-right) and j < (N-left):
            return True
        return False
    return False

def rho(i, j):
    if i == N//2:
        if j >= left and j < right:
            return 1.0
        elif j >= (N-right) and j < (N-left):
            return 1.0
        return 0.0
    return 0.0

def Efield(N, v):
    # Electric field
    # N-1 by N-1 matrix
    Esize = N - 1
    Ex = np.zeros((Esize, Esize), float)
    Ey = np.zeros((Esize, Esize), float)
    for i in range(Esize):
        for j in range(Esize):
            # Differential in two directions
            Ex[i][j] = (-0.5)*(v[i+1][j+1] - v[i+1][j] + v[i][j+1] - v[i][j])
            Ey[i][j] = (-0.5)*(v[i+1][j+1] - v[i][j+1] + v[i+1][j] - v[i][j])
    return [Ex, Ey]
```

```python
### times = 0
err = 1.0
n = 0
while err > 10**(-5):
    err = 0.0
    ierr = 0
    for i in range(1, N-1):
        for j in range(1, N-1):
            v0 = V[i, j]
            if not BC(i, j):
                V[i, j] = 0.25*(V[i+1][j] + V[i-1][j] + V[i][j+1] + V[i][j-1] + rho(i, j))
            err += np.abs(V[i, j] - v0)
            ierr += 1
    err /= ierr
    times += 1
    #print("times", times, ":", err)
    #er = np.append(er, verr)
    #print("it=", it[n], ":verr=", n)
print("Iteration times =", times)

print(V)
fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
Y = np.linspace(1, N, N)
X = np.linspace(1, N, N)
X, Y = np.meshgrid(X, Y)

ax1.contourf(X, Y, V)
#plt.colorbar()

E = Efield(N, V)
X = np.zeros(N-1, float)
Y = np.zeros(N-1, float)
for i in range(len(X)):
    X[i] = i+1.5
    Y[i] = i+1.5
X, Y = np.meshgrid(X, Y)
Ux = E[0]
Uy = E[1]
ax1.quiver(X, Y, Ux, Uy, scale = 7.4)

ax2 = fig.add_subplot(1, 2, 2)
x = np.zeros(N)
for i in range(N):
    x[i]=i
ax2.plot(x, V[17], 'b-')
#plt.savefig("hw2_2prime.png")
plt.show()
plt.plot(x, V[N//2], 'b-')
```
